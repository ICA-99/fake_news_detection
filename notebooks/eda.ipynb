Merge the Dataset

```python
df = pd.concat([df_fake_news, df_true_news], axis=0)
```

Shuffle Data

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df.head()
```

```python
print(df.info())
print(df["label"].value_counts())
```

```python
list(df["subject"].unique())
```

```python
df["content1"] = df["title"] + " " + df["text"]
```

```python
print(df.shape)
df.head(1)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
```

```python
fake_text = " ".join(df[df["label"] == 1]["content1"])
real_text = " ".join(df[df["label"] == 0]["content1"])
```

```python
fake_wc = WordCloud(
    width=800, 
    height=400, 
    background_color="white",
    stopwords=STOPWORDS
).generate(fake_text)

plt.figure(figsize=(10,5))
plt.imshow(fake_wc)
plt.axis("off")
plt.title("Fake News WordCloud")
plt.show()
```

```python
real_wc = WordCloud(
    width=800, 
    height=400, 
    background_color="white",
    stopwords=STOPWORDS
).generate(real_text)

plt.figure(figsize=(10,5))
plt.imshow(real_wc)
plt.axis("off")
plt.title("Real News WordCloud")
plt.show()
```
